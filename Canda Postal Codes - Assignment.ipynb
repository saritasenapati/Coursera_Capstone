```python
import requests
```

```python
Website_URL=requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
!pip install beautifulsoup4
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(Website_URL,'lxml')
print(soup.prettify())
```

```python
table = soup.find_all('table')[0]
```

```python
table
```

```python
import pandas as pd
df = pd.read_html(str(table))
```

```python
Canada_PostalCode_Df=df[0]
```

```python
Canada_PostalCode_Df.columns = Canada_PostalCode_Df.iloc[0]
Canada_PostalCode_Df=Canada_PostalCode_Df.drop(Canada_PostalCode_Df.index[0])

Canada_PostalCode_Df
```

```python
Canada_PostalCode_Df=Canada_PostalCode_Df[Canada_PostalCode_Df.Borough!='Not assigned']
Canada_PostalCode_Df
```

```python
Canada_PostalCode_Df.loc[Canada_PostalCode_Df.Neighbourhood == 'Not assigned', 'Neighbourhood'] = Canada_PostalCode_Df.Borough
Canada_PostalCode_Df
```

```python
Canada_PostalCode_Df=Canada_PostalCode_Df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(','.join).reset_index()
Canada_PostalCode_Df
```

```python
geocodes_df=pd.read_csv('C:\\Users\\Me\\Desktop\\Personal_Projects\\Coursera_Capstone\\Geospatial_Coordinates.csv')
geocodes_df.columns=['Postcode','Latitude','Longitude']
geocodes_df
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(Canada_PostalCode_Df['Borough'].unique()),
        Canada_PostalCode_Df.shape[0]
    )
)
```

```python
Canada_Postalcode_df_with_coordinates=Canada_PostalCode_Df.merge(geocodes_df,how='left',on='Postcode')
Canada_Postalcode_df_with_coordinates
```

```python
# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium
```

```python
!pip install geopy
from geopy.geocoders import Nominatim
address = 'Toronto'

geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronoto City are {}, {}.'.format(latitude, longitude))
```

```python
### Create a smaller subset by selecting only torono in bourogh
```

```python
Toronto_data = Canada_Postalcode_df_with_coordinates[Canada_Postalcode_df_with_coordinates['Borough'].str.contains("Toronto")].reset_index(drop=True)
Toronto_data
```

```python
# create map of New York using latitude and longitude values
map_Toronto= folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighbourhood in zip(Toronto_data['Latitude'], Toronto_data['Longitude'], Toronto_data['Borough'], Toronto_data['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)  
    
map_Toronto
```
